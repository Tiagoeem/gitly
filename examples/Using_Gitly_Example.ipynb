# Import, will install any dependences needed

````
from gitly.colab.plot import GitlyPlotter
````

# Simple example

````
from sklearn.datasets import make_blobs

# Creating a bunch of data 
n_samples = 6000
n_dimensions = 3
X, y = make_blobs(n_samples=n_samples, centers=6, n_features=n_dimensions, cluster_std=0.7, random_state=0)
````

````
import pandas as pd

# Just creating a dataframe with the fake data that we create
df = pd.DataFrame( data=X, columns=['x1', 'x2', 'x3'] )
df.head(3)
````

## Talk is cheap, show me the plot

````
# Instatiate gitly as a object of GitlyPlotter: Parameters for the constructor could be git, github or colab
gitly = GitlyPlotter('github')
````

This plot you will see in Github (is static), was created using Plotly

````
# Import plotly express to plot scatter 3D
import plotly.express as px

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig1 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig1 ) # Done!
````

## **This plot you will NOT see in Github**
You only can see if you run this notebook in Google Colab. But you can use all the tools and fency features of a Plotly Plot.

Just use:


```
gitly.config_render('colab')
```



````
# Change to dinamic plot in the same notebook
gitly.config_render('colab')

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig2 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig2 ) # Done!
````

Can I back to static? Of course!

Just use:


```
 gitly.config_render('github')
```


or


```
 gitly.config_render('git')
```

## Your plot will be static and will appers on Github.


````
# Change to static plot in the same notebook
gitly.config_render('git')

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig3 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig3 ) # Done!
````

# Change the size of the static plots

Is possible to change the height, width and scales of the statics plots.

To change the default values that you plot all graphs use the function ```config_render()```

```
gitly.config_render( 'git', default_height = 555, default_width = 444, default_scale = 0.7 )
```
After that all static plots will follow this configuration, dinamic standard Plotly plots will not change.


If you want only change the size of a specific plot use the arguments ```height```, ```width``` an ```scale``` in the ```.show()``` method.

```
gitly.show( fig, height = 88, scale = 3 ) # The width still the default
```


## Change the **default values**



````
# Change default values
gitly.config_render('git', default_scale = 0.25)
````

## Change only for an specific plot

Minimun value for **height** and **width** is **250**, if you need a tiny image let the height and width in the default value and use the scale parameter.

````
# the default still the scale 0.25, but will be overwrite
fig7 = px.scatter_3d( df, x='x2', y='x3', z='x1', color=y, title="Random Clusters")
gitly.show( fig7, format = 'jpg', height = 500,  width = 800, scale = 0.6 ) # Done!
````

# Import, will install any dependence if needed

````
from gitly.colab.plot import GitlyPlotter
````

# Simple example

````
from sklearn.datasets import make_blobs

# Creating a bunch of data 
n_samples = 6000
n_dimensions = 3
X, y = make_blobs(n_samples=n_samples, centers=6, n_features=n_dimensions, cluster_std=0.7, random_state=0)
````

````
import pandas as pd

# Just creating a dataframe with the fake data that we create
df = pd.DataFrame( data=X, columns=['x1', 'x2', 'x3'] )
df.head(3)
````

## Talk is cheap, show me the plot

````
# Instatiate gitly as a object of GitlyPlotter: Parameters for the constructor could be git, github or colab
gitly = GitlyPlotter('github')
````

This plot you will see in Github (is static), was created using Plotly

````
# Import plotly express to plot scatter 3D
import plotly.express as px

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig1 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig1 ) # Done!
````

This plot you will NOT see in Github, you only can see if you run this notebook in Google Colab. But you can use all the tools and fency features of a Plotly Plot.

Just use:


```
gitly.switch_renderer('colab')
```



````
# Change to dinamic plot in the same notebook
gitly.switch_renderer('colab')

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig2 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig2 ) # Done!
````

Can I back to static? Of course!

Just use:


```
# gitly.switch_renderer('github')
```


or


```
# gitly.switch_renderer('git')
```

And you plot will be static and will appers on Github.


````
# Change to static plot in the same notebook
gitly.switch_renderer('git')

# Create the figure as you usualy do, than instead of call fig.show(), pass fig to gitly.show( fig )
fig3 = px.scatter_3d( df, x='x1', y='x2', z='x3', color=y, title="Random Clusters")
gitly.show( fig3 ) # Done!
````
